```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Ejercicio_vslns.xlsx")
```

```python
g = sns.lmplot(x="PIBmlls", y= "MuertesCOVID", data= df, hue="Tiporegion")
plt.xlabel("PIB Mill $")
plt.ylabel("Muertes por COVID")
plt.show()
```

```python
g = sns.lmplot(x="PIBmlls", y= "MuertesCOVID", data= df, col="Tiporegion")
plt.xlabel("PIB Mill $")
plt.ylabel("Muertes por COVID")
plt.show()
```

```python
g = sns.jointplot(x="PIBmlls", y= "MuertesCOVID", data= df)
plt.xlabel("PIB Mill $")
plt.ylabel("Muertes por COVID")
plt.show()
```

```python
g = sns.jointplot(x="PIBmlls", y= "MuertesCOVID", data= df, kind="reg")
plt.xlabel("PIB Mill $")
plt.ylabel("Muertes por COVID")
plt.show()
```

```python
g = sns.jointplot(x="PIBmlls", y= "MuertesCOVID", data= df, kind="hex")
plt.xlabel("PIB Mill $")
plt.ylabel("Muertes por COVID")
plt.show()
```

```python
g = sns.jointplot(x="PIBmlls", y= "MuertesCOVID", data= df, kind="kde")
plt.xlabel("PIB Mill $")
plt.ylabel("Muertes por COVID")
plt.show()
```
